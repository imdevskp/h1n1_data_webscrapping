## Africa

https://en.wikipedia.org/wiki/2009_swine_flu_pandemic_in_Africa

```python
# link
link = 'https://en.wikipedia.org/wiki/2009_swine_flu_pandemic_in_Africa'
# get data
tables = pd.read_html(link)
# no. of tables
len(tables)
```

```python
df = tables[0]
df = df.drop([0, 37, 38, 39])
df.columns = ['Country', 'Cumulative no. of lab confirmed cases', 'Cumulative no. confirmed deaths']
for col in df.columns[1:]:
    df[col] = df[col].str.replace(',', '')
    df[col] = df[col].str.extract('(\d+)')
    df[col] = df[col].astype('int')
df = df.sort_values('Cumulative no. of lab confirmed cases', ascending=False)\
        .reset_index(drop=True)
df.to_csv('summary_africa.csv', index=False)
df
```
